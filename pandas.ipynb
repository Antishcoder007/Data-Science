```python
data = {'A':[2,4,5,np.nan,78,np.nan,8],
       'B':[12,14,np.nan,78,np.nan,8,55],
       'C':[np.nan,4,5,78,np.nan,8,63],
        'D':[2,4,5,500,78,11,8]}
data
```

## fill the value of missing placeses

```python
# missing values imputation
# difference between 'Nan' and '0'
# Nan is a blank space which can hold value 
# 0 itself a value which is take place
```

```python
# to fill the value in empty space you have to fill by column approches
# numeracal :- constant valiue , programing  or manually
# mean , median value fill
```

```python
sample
```

```python
# to fill consatant value in A field

sample['A'].fillna(0)
```

```python
sample
```

```python
# to fill constant value in whole data set 
# sample.fillna(10):- it will fill the data in whole data set wherer the Nan field is avilabele
sample.fillna(10)
```

```python
sample['A'].mean()
```

```python
sample['A'].median()
```

```python
sample['A'].fillna(sample['A'].mean())
```

```python
sample['A'].fillna(sample['A'].median())
```

```python
# when we will use mean and median
# we will check the data is normally distributed or not
# if the data is normal distrubuted we will go with mean 
# and is the data is not normal distributed we will go with median
```
